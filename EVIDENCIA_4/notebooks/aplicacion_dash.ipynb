```python
from dash import Dash, dcc, html
from dash.dependencies import Output, Input


app = Dash(__name__)

app.layout = html.Div((
    "Pick a Color!: ", 
    dcc.Dropdown(
        id="color-input",
         options=["Rojo", "Amarillo", "Azul", "Verde", "Negro"],
         value = "Amarillo"
    ),
))
@app.callback(
    Output("color-output", "children"),
    Input("color-input", "value")
)
def update_output_div(color):
    return f"Color Seleccionado: {color}"

if __name__ == '__main__':
    app.run()
```

```python
from dash import Dash, html, dcc
from dash.dependencies import Output, Input
from dash.exceptions import PreventUpdate
app = Dash(__name__)

app.layout = html.Div(
    ["Selecciona un color: ",
     dcc.Dropdown(
         id="color-input",
         options=["Rojo", "Amarillo", "Azul", "Verde", "Negro"],
         value = "Amarillo"
     ),
     html.Div(id="color-output")
     ])

@app.callback(
    Output("color-output", "children"),
    Input("color-input", "value")
)
def update_output_div(color):
    return f"Color Seleccionado: {color}"


if __name__ == "_main_":
    app.run(mode="inline", debug= True,port="8051")
```
